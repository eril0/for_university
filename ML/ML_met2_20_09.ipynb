```python
#7
def replace_func(x):
    if x<63:
        return 2
    elif x>62 and x<77:
        return 3
    elif x>76 and x<91:
        return 4
    else:
        return 5
```

```python
df['Балл 5'] = df.loc[:,'Балл']
df
```

```python
df['Балл 5'] = df['Балл 5'].apply(replace_func)
df
```

```python
df.groupby(['Балл 5']).agg({'Балл 5':'count'})/df.shape[0]*100
```

```python
#10
kor=len(df.loc[0,'Задания с кратким ответом'])
razv=int(len(df.loc[0,'Задания с развёрнутым ответом'])/4)
kor,razv
```

```python
#11
for i in range(kor):
    df.insert(len(df.columns),f'B{i+1}',0)
df
```

```python
for i,row in df.iterrows():
    kor=list(df.loc[i,'Задания с кратким ответом'])
    for j in range(len(kor)):
        df.loc[i,f'B{j+1}']=kor[j]

df
```

```python
cols=df.loc[:,'B1':'B12'].columns
for col in cols:
    print(df.groupby([col]).agg({col:'count'})/df.shape[0]*100)
```

```python
#12
for i in range(razv):
    df.insert(len(df.columns),f'C{i+1}',0)
df
```

```python
for i,row in df.iterrows():
    razv=list(df.loc[i,'Задания с развёрнутым ответом'])
    razv1=[]
    for s in range(0,len(razv),4):
        razv1.append(razv[s])
    for j in range(len(razv1)):
        df.loc[i,f'C{j+1}']=razv1[j]

df
```

```python
cols=df.loc[:,'C1':'C7'].columns
for col in cols:
    print(df.groupby([col]).agg({col:'count'})/df.shape[0]*100)
```
